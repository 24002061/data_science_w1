# List

```python
list1=[3,12,11,14,9] #how you do list
print(list1, list1[2])
```

```python
list2=[1,4,3,9]
print(list2[0]) #the index 1=0 ,4=1 ,3=2 (just like the binary number)
print(list2[3])
print(list2[-1]) #the index from the back use -ve
```

```python
print(list2)
```

```python
list1=[3,12,11] #how you do list
list1[2]='data' #replacing the value present at index 3
print(list1)
```

```python
list1.append('Science')
print(list1)
#the size of the list is flexible
```

```python
list1.pop(1) #delete element
print(list1)
```

```python
numbers=list(range(6))
print(numbers)
```

```python
#slicing
list2=[11,2,5,18,3,20]
print(list2[2:4]) # value at index 2 & 3 only (not teh fourth one)
print(list2[2:]) #print all number index 2
print(list2[:2]) #print every numbers before index 2
print(list2[:-1])
list2[2:4]=[50,100] # replacing the previous list
print(list2)
```

# Dictionary

```python
d={'cat': 'cute', 'dog': 'furry'} #key-value, key can be of any type
print(d['cat']) # cat=key ; dog=key
print('cat' in d)
print('catt' in d)
```

```python
d['dog']='not furry' # cannot change the key
print(d)
```

```python
del d['cat'] # delete the key
print(d)
```
