```python
import numpy as np
```

```python
#Create a numpy array of size 2 x 2
a=np.array([[2,2],[3,3]])
print(a)
```

```python
#Create a random numpy array of size 3 x 4
b=np.random.rand(3,4)
print(b)
```

```python
#Create a numpy array of ones and a numpy array of zeroes
c=np.ones((2,2))
d=np.zeros((2,2))

print("This is C")
print(c)
print("This is D")
print(d)
```

```python
#Create a random numpy array of size 6x3 and transpose it
e=np.random.rand(6,3)
print(e)

print("After transpose")
print(e.T)
```

```python
#Create two random numpy array of size n x p and p x m and then compute the dot product
f=np.random.rand(3,2)
g=np.random.rand(2,1)

print("This is F")
print(f)
print("This is G")
print(g)

print("Dot product")
print(np.dot(f,g))
```
